```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[1,2,7,2,1,4,5,6]
y2=[1,2,5,4,2,4,5,6]
y3=[1,2,3,1,5,2,6,4]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('plot graph',fontsize=24)
ax.set_xlabel('<............x axis...........>',fontsize=14)
ax.set_ylabel('<............y axis............>',fontsize=20)
ax.legend()
plt.plot(x,y,'-..r',label='curve1')
plt.plot(x,y2,'--y',label='curve2')
plt.plot(x,y3,'--m',label='curve3')
ax.legend()
```
