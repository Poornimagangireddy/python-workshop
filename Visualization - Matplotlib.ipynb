# Matplotlib for Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

## Simple Examples

x= [1,2,3,4,5]
y=[6,7,8,9,3]
```

```python
##plotting graphs using matplotlib 

##scatter plot

plt.scatter(x,y,c ='g')
plt.xlabel('poori')
plt.ylabel('puppy')
plt.title('sisters')
plt.savefig('Test.png')
```

```python
## plt plot

plt.plot(x,y,'g*',linestyle='--',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 12.5, 0.1) 

y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 12, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a sub
#plot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g-') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
# Histograms
a = np.array([10,4,6,6,7,5,5,7,10,3]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
#Boxplot
value1 = [82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[6,5,91,25,36,32,96,95,3,90,95,32,27,55,100,15,71,11,37,21]
value3=[23,89,12,78,72,89,25,69,68,86,19,1,15,16,16,75,65,31,25,52]
value4=[59,73,70,16,81,1,88,98,10,87,29,72,16,23,72,88,78,99,75,30]
 
box=[value1,value2,value3,value4]
plt.boxplot(box)
plt.show()
```

```python
#adding colours
box=[value1,value2,value3,value4]
plt.boxplot(box,patch_artist=True,labels=['course1','course2','course3','course4'])
plt.show()
```

```python
#horizontal boxplot
box_plot_data=[value1,value2,value3,value4]
box=plt.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['course1','course2','course3','course4'],)
```

```python
#Pie chart
marks = [215, 130, 245, 210]
labels = 'Python', 'C++', 'Ruby', 'Java'
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.2, 0, 0)  # explode 1st slice

# Plot
plt.pie(marks, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```
